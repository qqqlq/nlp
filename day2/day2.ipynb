```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    "私は犬が好き",
    "私は猫が好き",
    "犬も猫も好き",
]

# 文字単位でBoWを作成（analyzer="char"の意味は後で説明）
vectorizer = CountVectorizer(analyzer="char")
X = vectorizer.fit_transform(corpus)

print("単語一覧:", vectorizer.get_feature_names_out())
print("\n行列:\n", X.toarray())
```

```python
# 英語ならスペース区切りで自然に単語分割できる
corpus_en = [
    "I like dogs",
    "I like cats",
    "dogs and cats",
]

vectorizer_en = CountVectorizer()  # デフォルトはスペース区切り
X_en = vectorizer_en.fit_transform(corpus_en)

print("単語一覧:", vectorizer_en.get_feature_names_out())
print("\n行列:\n", X_en.toarray())
```

```python
# BoWの問題点を見る
corpus2 = [
    "犬が好き 犬は かわいい 犬を 飼いたい",
    "猫が好き 猫は かわいい",
    "犬も 猫も かわいい",
]

vectorizer2 = CountVectorizer(analyzer="char")
X2 = vectorizer2.fit_transform(corpus2)

print("単語一覧:", vectorizer2.get_feature_names_out())
print("\n行列:\n", X2.toarray())
```

TF  = その単語がその文書に何回出てくるか（多いほど重要）
IDF = その単語が全文書のうち何割に出てくるか（少ない文書にしか出ない単語ほど重要）

TF-IDF = TF × IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# スペース区切りの英語で試す（日本語は後でちゃんとやる）
corpus_en2 = [
    "I like dogs dogs dogs",  # dogsが3回
    "I like cats",
    "dogs and cats are cute",
]

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(corpus_en2)

print("単語一覧:", tfidf.get_feature_names_out())
print("\n行列(小数第2位まで):\n", X_tfidf.toarray().round(2))
```

```python
corpus = [
    "I like dogs dogs dogs",
    "I like cats",
    "dogs and cats are cute",
]

vectorizer = CountVectorizer()

# fitとtransformを分けてやってみる
vectorizer.fit(corpus)
print("単語一覧:", vectorizer.get_feature_names_out())

X = vectorizer.transform(corpus)
print("\n疎行列:\n", X)
print("\nNumPy配列:\n", X.toarray())
```
