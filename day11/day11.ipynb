```python
from transformers import pipeline

# 感情分析のパイプラインを作る（初回はモデルをダウンロード）
print("モデルをロード中...")
classifier = pipeline("sentiment-analysis")
print("完了!")

# 試してみる
result = classifier("I love this movie! It was fantastic.")
print(result)
```

```python
# 複数の文章を一度に処理する
texts = [
    "I love this movie! It was fantastic.",
    "This film was terrible and boring.",
    "The acting was okay but the story was confusing.",
    "Absolutely amazing! Best movie I have ever seen.",
]

results = classifier(texts)

for text, result in zip(texts, results):
    label = result['label']
    score = result['score']
    print(f"{label} ({score:.3f}): {text[:40]}...")
```

```python
# pipelineが内部で何を使っているか確認する
print("モデル名:", classifier.model.config.name_or_path)
print("モデルの種類:", type(classifier.model))
print("\nモデルの構造（一部）:")
print(classifier.model.config)
```

```python
from transformers import AutoTokenizer

# トークナイザーをロード
tokenizer = AutoTokenizer.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)

# テキストをトークン化
text = "I love dogs and cats!"
tokens = tokenizer(text)
print("トークンID:", tokens['input_ids'])
print("トークン文字列:", tokenizer.convert_ids_to_tokens(tokens['input_ids']))
```

```python
# 長い単語や珍しい単語はどうなるか
text2 = "The movie was unbelievably phenomenal!"
tokens2 = tokenizer(text2)
print("トークン文字列:", tokenizer.convert_ids_to_tokens(tokens2['input_ids']))
```

```python
words = ["plane", "plan", "planet", "planning", "unbelievably", "unbelievable"]
for word in words:
    tokens = tokenizer.tokenize(word)
    print(f"{word:20} → {tokens}")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# トークナイザーとモデルを別々にロードする
tokenizer = AutoTokenizer.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)
model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)

# テキストをトークン化する
text = "I love dogs and cats!"
inputs = tokenizer(text, return_tensors="pt")
print("入力の形:", inputs['input_ids'].shape)
print("入力トークンID:", inputs['input_ids'])
```

```python
# モデルに通す
with torch.no_grad():  # 推論時は勾配計算不要
    outputs = model(**inputs)

print("出力の型:", type(outputs))
print("logitsの形:", outputs.logits.shape)
print("logits:", outputs.logits)

# logitsをsoftmaxで確率に変換する
probs = torch.softmax(outputs.logits, dim=-1)
print("\n確率:", probs)
print("NEGATIVE:", probs[0][0].item())
print("POSITIVE:", probs[0][1].item())

print("NEGATIVE:", probs[0][0])
```
