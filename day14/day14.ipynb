```python
from transformers import MarianTokenizer, MarianMTModel

# Marianの専用クラスでロードする
model_name = "Helsinki-NLP/opus-mt-ja-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model.generate(**inputs)
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result

texts = [
    "私は犬が好きです。",
    "今日の天気はとても良いですね。",
    "機械学習はとても面白い分野です。",
]

for text in texts:
    print(f"原文：{text}")
    print(f"翻訳：{translate(text)}")
    print()
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ja-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-ja-en")

def translate(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model.generate(**inputs)
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result

texts = [
    "私は犬が好きです。",
    "今日の天気はとても良いですね。",
    "機械学習はとても面白い分野です。",
]

for text in texts:
    print(f"原文：{text}")
    print(f"翻訳：{translate(text)}")
    print()
```

```python
model_name_en_ja = "Helsinki-NLP/opus-mt-en-jap"
tokenizer_en_ja = AutoTokenizer.from_pretrained(model_name_en_ja)
model_en_ja = AutoModelForSeq2SeqLM.from_pretrained(model_name_en_ja)

def translate_en_ja(text):
    inputs = tokenizer_en_ja(text, return_tensors="pt")
    outputs = model_en_ja.generate(**inputs)
    return tokenizer_en_ja.decode(outputs[0], skip_special_tokens=True)

texts_en = [
    "I like dogs.",
    "Machine learning is a very interesting field.",
    "The weather is very nice today.",
]

for text in texts_en:
    print(f"原文：{text}")
    print(f"翻訳：{translate_en_ja(text)}")
    print()
```

```python
# BLEUスコアで翻訳精度を評価する
# pip install sacrebleu が必要
import subprocess
subprocess.run(["pip", "install", "sacrebleu", "--quiet"])

import sacrebleu

# 参照翻訳（正解）と機械翻訳の結果を比較する
references = [
    ["I like dogs."],
    ["The weather is very nice today, isn't it?"],
    ["Machine learning is a very interesting field."],
]

hypotheses = [
    translate("私は犬が好きです。"),
    translate("今日の天気はとても良いですね。"),
    translate("機械学習はとても面白い分野です。"),
]

print("機械翻訳の結果:")
for hyp, ref in zip(hypotheses, references):
    print(f"  翻訳：{hyp}")
    print(f"  参照：{ref[0]}")
    print()

# BLEUスコアを計算する
bleu = sacrebleu.corpus_bleu(hypotheses, list(map(list, zip(*references))))
print(f"BLEUスコア: {bleu.score:.2f}")
```
