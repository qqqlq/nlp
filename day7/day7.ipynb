```python
import torch
import torch.nn as nn

# nn.Moduleを使ってネットワークを定義する
class SimpleNN(nn.Module): # nn.Moduleを継承してクラスを定義
    def __init__(self): # インスタンス化時に自動実行される初期化メソッド
        super().__init__() # nn.Mosduleの初期化を先に実行　# 親クラスの初期化
        # 層を定義する　# その上に自分の初期化を追加　
        self.layer1 = nn.Linear(2, 3) # 入力が2次元、出力が3次元
        self.layer2 = nn.Linear(3, 1) # 入力が3次元、出力が1次元
        
    def forward(self, x):
        # 順伝播の計算を定義する
        h = torch.relu(self.layer1(x)) # 活性化関数ReLUを適用
        y = torch.sigmoid(self.layer2(h)) # 活性化関数Sigmoidを適用
        return y
    
# ネットワークを作成
model = SimpleNN() # __init__が呼ばれる → layer1とlayer2が定義される
print(model) # self.layer1とself.layer2の情報が表示される
```

```python
x = torch.tensor([1.0, 2.0])

# model(x) で順伝播
y_pred = model(x)
print("予測値:", y_pred)
print("形:", y_pred.shape)
```

```python
# optimizerを使った学習ループ
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) # モデルのパラメータを最適化するためのオプティマイザを定義
loss_fn = nn.BCELoss() # 二値分類の損失関数を定義

x = torch.tensor([1.0, 2.0])
y_true = torch.tensor([1.0])

losses = []

for i in range(100):
    # 順伝播
    y_pred = model(x)
    loss = loss_fn(y_pred, y_true) # y_trueの形をy_predに合わせる
    
    # 逆伝播
    optimizer.zero_grad() # 勾配を初期化
    loss.backward() # 勾配を計算
    optimizer.step() # パラメータを更新
    
    losses.append(loss.item()) # 損失をリストに保存
    
print("初期損失:", losses[0])
print("最終損失:", losses[-1])
```
