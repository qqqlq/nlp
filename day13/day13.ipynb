```python
from transformers import pipeline

# GPT-2でテキスト生成
generator = pipeline("text-generation", model="gpt2")

result = generator(
    "The future of artificial intelligence",
    max_new_tokens=50,
    num_return_sequences=1,
)

print("Generated Text:")
print(result[0]['generated_text'])
```

```python
from transformers import set_seed

prompt = "The future of artificial intelligence"

for temp in [0.1, 0.7, 1.5]:
    result = generator(
        prompt,
        max_new_tokens=30,
        temperature=temp,
        do_sample=True,
    )
    print(f"Temperature: {temp}")
    print(result[0]['generated_text'])
    print()
```

```python
# temperatureとは
# softmaxの前にスコアを割り算する値。
# temperature=0.1（低い）：確率が最も高いトークンに集中する → 無難・繰り返しが多い
# temperature=1.0（標準）：通常のsoftmax
# temperature=1.5（高い）：確率が均等になる → 多様だが意味が崩れやすい


import numpy as np

scores = np.array([3.0, 1.0, 0.5])
print("Original scores:", scores)

for temp in [0.1, 1.0, 1.5]:
    print("---------------------------------")
    scaled = scores / temp          # ① temperatureで割る
    exp = np.exp(scaled)            # ② expをとる
    probs = exp / exp.sum()         # ③ 合計で割る（softmax）
    print(f"scaled scores (temp={temp}): {scaled.round(3)}")
    print(f"exp(scaled scores): {exp.round(3)}")
    print(f"temperature={temp}: {probs.round(3)}")
```

```python
# top_k: 上位k個のトークンからランダムに選ぶ
result_topk = generator(
    prompt,
    max_new_tokens=30,
    do_sample=True,
    top_k=10, # 生成候補を上位10個に絞る
)
print("top_k=10:")
print(result_topk[0]['generated_text'])
print()

# top_p: 累積確率がp以上になるまでトークンを選ぶ
result_topp = generator(
    prompt,
    max_new_tokens=30,
    do_sample=True,
    top_p=0.9, # 累積確率が90%になるまでトークンを選ぶ
)
print("top_p=0.9:")
print(result_topp[0]['generated_text'])
print()
```
